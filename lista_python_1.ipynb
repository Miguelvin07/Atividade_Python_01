## Questao 9
Faça um Programa que peça 2 números inteiros e um número real. Calcule e mostre:
- o produto do dobro do primeiro com metade do segundo.
- a soma do triplo do primeiro com o terceiro.
- o terceiro elevado ao cubo.

```python
inteiro1 = int(input("Digite o primeiro número inteiro" ))
inteiro2 = int(input("Digite o segundo número inteiro" ))
real = float(input("Digite um número real" ))

produto = (2*inteiro1) * (inteiro2/2)
soma = (3*inteiro1) + real
expoente = real**3

print(f"Os seus resultados são: {produto}, {soma}, {expoente}")
```

## Questão 11

Faça um programa que calcule o resultado da seguinte equação

$$ 2 \bigg[ \frac{3\times(2+5)-20}{3} \bigg]^2 + \sqrt 2 + 1  $$

## Questão 12

 Para valores de $x$ em $\{ 1,5,8, 9\}$, compute o valor da função $\quad{}f(x) = \beta_0 + \beta_1 x \quad{}$, com parâmetros $\beta_0 = -1.3$ e $\beta_1=3.5$.

```python
b0 = -1.3
b1 = 3.5
valores = [1,5,8,9]
for x in valores:
  fx = b0 + b1*x
  print(f"Os valores de x são : F(x)={fx}")
```
